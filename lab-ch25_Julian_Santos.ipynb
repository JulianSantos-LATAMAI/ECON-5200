### DIAGNOSE: Error 1 — Chunking Configuration

The text splitter below has **two problems** with its configuration. Run the cell and examine the output to identify them.

````python
# -----------------------------------------------------------
# DIAGNOSE: This code has errors in the chunking configuration.
# -----------------------------------------------------------

# ERROR 1a: chunk_size is far too small — 50 characters means each
# chunk is just a few words, destroying semantic coherence.
# A good chunk_size for RAG is typically 500-1000 characters.
#
# ERROR 1b: chunk_overlap is 0 — sentences at chunk boundaries
# get split in the middle, losing context. A typical overlap
# is 10-20% of chunk_size (e.g., 100 for chunk_size=500).

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=50,        # BUG: way too small!
    chunk_overlap=0,      # BUG: no overlap means lost context at boundaries
    length_function=len,
    separators=["\n\n", "\n", ". ", " ", ""]
)

chunks = text_splitter.split_documents(fomc_documents)

print(f'Number of chunks: {len(chunks)}')
print(f'Average chunk length: {np.mean([len(c.page_content) for c in chunks]):.0f} characters')
print(f'\nFirst 5 chunks:')
for i, chunk in enumerate(chunks[:5]):
    print(f'  Chunk {i}: "{chunk.page_content[:80]}..." ({len(chunk.page_content)} chars)')

print(f'\nPROBLEM: With 50-character chunks, each chunk is just a sentence fragment.')
print('The retriever will match on fragments, not meaningful passages.')
````

````python
text_splitter_fixed = RecursiveCharacterTextSplitter(
    chunk_size=700,
    chunk_overlap=100,
    length_function=len,
    separators=["\n\n", "\n", ". ", " ", ""]
)

chunks_fixed = text_splitter_fixed.split_documents(fomc_documents)

print(f'Number of chunks: {len(chunks_fixed)}')
print(f'Average chunk length: {np.mean([len(c.page_content) for c in chunks_fixed]):.0f} characters')
print(f'\nFirst 3 chunks:')
for i, chunk in enumerate(chunks_fixed[:3]):
    print(f'  Chunk {i}: "{chunk.page_content[:120]}" ({len(chunk.page_content)} chars)')
````

Answer each question in 2-3 sentences.

### Q1: When does RAG outperform direct LLM prompting? When does it not matter?

RAG outperforms direct prompting when the answer depends on specific documents the model was never trained on — like proprietary FOMC minutes, internal company reports, or recent regulatory filings — because it grounds responses in retrieved text rather than relying on the model's memory. As seen in Q2, RAG was clearly superior for tracing the Fed's stance across specific meetings, while for Q2 (the rate target) it didn't matter much since the fact was already in the model's training data.



### Q2: How do chunking parameters affect retrieval quality?

*Connect to the bias-variance tradeoff from Ch 15 — small chunks have high variance (noisy retrieval), large chunks have high bias (diluted relevance).*
Chunking creates a bias-variance tradeoff similar to the one from Ch 15: very small chunks like the buggy 50-character setting produce high-variance retrieval where fragments are too noisy and lack enough context to be meaningful, similar to an overfit model reacting to noise. Conversely, chunks that are too large introduce high bias by diluting the relevant signal across too much irrelevant text, like an underfit model averaging over too much data — the sweet spot of 700 characters with 100-character overlap balanced both problems.




### Q3: In what economic research contexts would you use RAG over direct prompting?

*Think about: policy analysis on proprietary documents, literature review, regulatory compliance, central bank communication analysis.*

RAG is most valuable in economic research when working with proprietary or time-sensitive documents that never entered the model's training data, such as internal central bank communications, confidential regulatory filings, or unpublished working papers. It's also essential for systematic literature review or policy analysis where every claim needs to be traceable to a specific source — for example, analyzing how the Fed's language around inflation shifted across FOMC meetings, exactly as we did in this lab.
